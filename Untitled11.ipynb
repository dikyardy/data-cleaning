```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import seaborn as sns
import time
import re
from wordcloud import WordCloud
%matplotlib inline
import plotly.express as px
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df = pd.read_csv("dst.csv", sep=";")
```

```python
df.head()
```

```python
import string
string.punctuation
```

```python
def remove_punctuation(txt):
  txt_nopunct = "".join([c for c in txt if c not in string.punctuation])
  return txt_nopunct
```

```python
df['text_clean'] = df['text'].apply(lambda x: remove_punctuation(x))
df.head()
```

```python
import re

def tokenize(txt):
  tokens = re.split('\W+', txt)
  return tokens
```

```python
df['text_clean_tokenized'] = df['text_clean'].apply(lambda x: tokenize(x.lower()))
df.head()
```

```python
import nltk
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('indonesian')
```

```python
def remove_stopwords(txt_tokenized):
  txt_clean =[word for word in txt_tokenized if word not in stopword]
  return txt_clean
```

```python
df['text_no_sw'] = df['text_clean_tokenized'].apply(lambda x: remove_stopwords(x))
df.head()
```

```python
#### STEMMING

import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
dir(ps)
```

```python
def stemming(tokenized_text):
  text = [ps.stem(word) for word in tokenized_text]
  return text
```

```python
df['text_stmmd'] = df['text_no_sw'].apply(lambda x : stemming(x))
df.head()
```

```python
### LEMMATIZATION
from nltk.stem.wordnet import WordNetLemmatizer
import nltk
wn = WordNetLemmatizer
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('indonesian')
dir(wn)
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
#### NOW LEMMATIZE

import nltk
wn = WordNetLemmatizer()

def lemmatization(token_txt):
  text =[wn.lemmatize (word) for word in token_txt]
  return text
```

```python
df['text_lemm'] = df['text_stmmd'].apply(lambda x: lemmatization(x))
df.head()
```

```python
df.head()
```

```python
# Manual encoding of labels

label_dict = {'normal': 0, 'hate-speech': 1}
df.replace({'kategori': label_dict}, inplace = True)

print(pd.Series({"Memory usage": "{:.2f} MB".format(df.memory_usage().sum()/(1024*1024)),
                 "Dataset shape": "{}".format(df.shape)}).to_string())
```

```python
# Splitting the dataset by label
data_n = df[df['kategori'] == 0] # normal
data_p = df[df['kategori'] == 1] # hate-speech
```

```python
# Visualization of class frequencies
import plotly.graph_objects as go
values = np.array([len(data_n), len(data_p)])
labels = ['normal', 'hate-speech',]
fig = go.Figure(data = [go.Pie(values = values, labels = labels, hole = 0.5, textinfo = 'percent', title = " ")])
text_title = "Comparison of class frequencies"
fig.update_layout(height = 500, width = 800, showlegend = True, title = dict(text = text_title, x = 0.5, y = 0.95)) 
fig.show()
```
